```python
# ==========================================================
#  Stable Diffusion WebUI (A1111) ã‚’ Python 3.10 ã§èµ·å‹•ã™ã‚‹ç‰ˆ (ControlNetãªã—)
#  - Colab ã‚«ãƒ¼ãƒãƒ«ã¯ 3.11/3.12 ã§ã‚‚OKï¼ˆå®Ÿè¡Œç³»ã ã‘ 3.10ï¼‰
#  - xFormers æœªä½¿ç”¨ï¼ˆ--opt-sdp-attention ã§SDPAï¼‰
#  - Civitai å–å¾—ãƒ­ã‚¸ãƒƒã‚¯ã‚’è¸è¥²
# ==========================================================

print("ğŸ§ª Notebook kernel Python (å‚è€ƒ):")
!python -V

import os, sys, json, shutil, subprocess, time
from google.colab import drive, userdata

# ----------------------------
# 0) Micromamba ã§ Python 3.10 ç’°å¢ƒã‚’ç”¨æ„
# ----------------------------
print("\nğŸ“¦ micromamba ã‚’ã‚»ãƒƒãƒˆã‚¢ãƒƒãƒ—ã—ã¦ py310 ç’°å¢ƒã‚’ä½œæˆ...")
MICROMAMBA = "/usr/local/bin/micromamba"
if not os.path.exists(MICROMAMBA):
    # å®‰å®šã—ã¦ä½¿ã‚ã‚Œã‚‹å…¥æ‰‹æ³•ï¼ˆå…¬å¼é…å¸ƒã‚¢ãƒ¼ã‚«ã‚¤ãƒ–ï¼‰
    !wget -qO /tmp/micromamba.tar.bz2 https://micromamba.snakepit.net/api/micromamba/linux-64/latest
    !tar -xjf /tmp/micromamba.tar.bz2 -C /usr/local/bin bin/micromamba --strip-components=1

# py310 ç’°å¢ƒã‚’ä½œæˆï¼ˆå­˜åœ¨ã™ã‚Œã°ã‚¹ã‚­ãƒƒãƒ—ï¼‰
!micromamba create -y -n py310 -c conda-forge python=3.10.13 pip git aria2 curl

print("\nğŸ§ª Runtime (py310) ãƒãƒ¼ã‚¸ãƒ§ãƒ³ç¢ºèªï¼š")
!micromamba run -n py310 python -V

# ----------------------------
# 1) ã‚·ã‚¹ãƒ†ãƒ å´ã®æœ€ä½é™ãƒ„ãƒ¼ãƒ« & Drive
# ----------------------------
print("\nğŸ”§ apt / Google Drive æº–å‚™...")
!apt-get update -qq
!apt-get install -y -qq aria2 wget curl

try:
    drive.mount('/content/drive')
    print("âœ… Google Drive ã‚’ãƒã‚¦ãƒ³ãƒˆã—ã¾ã—ãŸ")
except Exception as e:
    print(f"âš ï¸ Drive ãƒã‚¦ãƒ³ãƒˆå¤±æ•—ã¾ãŸã¯æ—¢ã«ãƒã‚¦ãƒ³ãƒˆæ¸ˆã¿: {e}")

drive_base_data_path = '/content/drive/MyDrive/sd_colab_data'
drive_models_path     = os.path.join(drive_base_data_path, 'models/Stable-diffusion')
drive_vae_path        = os.path.join(drive_base_data_path, 'models/VAE-approx')
os.makedirs(drive_models_path, exist_ok=True)
os.makedirs(drive_vae_path, exist_ok=True)

# ----------------------------
# 2) æ—¢å­˜WebUIå‰Šé™¤ â†’ ã‚¯ãƒ­ãƒ¼ãƒ³
# ----------------------------
webui_path = '/content/stable-diffusion-webui'
if os.path.exists(webui_path):
    shutil.rmtree(webui_path)
    print(f"ğŸ§¹ æ—¢å­˜ã®WebUIã‚’å‰Šé™¤: {webui_path}")

print("\nâ¬‡ï¸ Stable Diffusion WebUI ã‚’ã‚¯ãƒ­ãƒ¼ãƒ³...")
%cd /content
!git clone -q https://github.com/AUTOMATIC1111/stable-diffusion-webui.git

# ----------------------------
# 3) PyTorch (cu121) ã‚’ã€Œpy310ç’°å¢ƒã€ã«ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«
# ----------------------------
print("\nğŸ”¥ PyTorch (cu121) ã‚’ py310 ç’°å¢ƒã¸ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«...")
!micromamba run -n py310 python -m pip -q install --upgrade pip
!micromamba run -n py310 pip install -q --index-url https://download.pytorch.org/whl/cu121 torch torchvision torchaudio

# è¿½åŠ ãƒ©ã‚¤ãƒ–ãƒ©ãƒªï¼ˆWebUIã‚„SDXLã§ã‚ˆãä½¿ã†ã‚„ã¤ï¼‰ï¼špy310 ç’°å¢ƒã¸
print("\nğŸ“š è¿½åŠ ãƒ©ã‚¤ãƒ–ãƒ©ãƒªï¼ˆpy310 ç’°å¢ƒï¼‰ã‚’ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«...")
!micromamba run -n py310 pip install -q open_clip_torch transformers timm requests

# ----------------------------
# 4) æ‹¡å¼µï¼ˆCivitai Helper ã®ã¿ï¼‰
# ----------------------------
print("\nğŸ”Œ æ‹¡å¼µæ©Ÿèƒ½ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«...")
%cd /content/stable-diffusion-webui/extensions
!git clone -q https://github.com/butaixianran/Stable-Diffusion-Webui-Civitai-Helper.git
print("âœ… æ‹¡å¼µã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«å®Œäº†")

# ----------------------------
# 5) Civitai API ãƒˆãƒ¼ã‚¯ãƒ³ & ãƒ¢ãƒ‡ãƒ«å–å¾—
# ----------------------------
print("\nğŸ”‘ Civitai API ãƒˆãƒ¼ã‚¯ãƒ³ã‚’å–å¾—ï¼ˆColab å·¦ã®ğŸ”‘ã‹ã‚‰äº‹å‰ç™»éŒ²ï¼‰...")
try:
    CIVI = userdata.get('CIVITAI_API_TOKEN')
    if not CIVI or not CIVI.strip():
        raise ValueError("CIVITAI_API_TOKEN ãŒæœªè¨­å®š")
    os.environ['CIVITAI_API_TOKEN'] = CIVI.strip()
    print("âœ… CIVITAI_API_TOKEN ã‚’æ¤œå‡º")
except Exception as e:
    print("âŒ Civitai APIãƒˆãƒ¼ã‚¯ãƒ³ãŒè¦‹ã¤ã‹ã‚Šã¾ã›ã‚“")
    print("   â†’ ã‚µã‚¤ãƒ‰ãƒãƒ¼ã® ğŸ”‘ Secrets ã§ `CIVITAI_API_TOKEN` ã‚’è¨­å®šã—ã¦å†å®Ÿè¡Œã—ã¦ãã ã•ã„ã€‚")
    raise

def get_latest_model_version_id(token: str):
    print("ğŸ” WAI-NSFW-illustrious-SDXL ã®æœ€æ–°ãƒãƒ¼ã‚¸ãƒ§ãƒ³IDã‚’å–å¾—ä¸­...")
    model_id = "827184"  # WAI-NSFW-illustrious-SDXL
    api_url  = f"https://civitai.com/api/v1/models/{model_id}?token={token}"
    r = subprocess.run(['curl','-s',api_url], capture_output=True, text=True)
    if r.returncode != 0 or not r.stdout:
        print("âš ï¸ APIå‘¼ã³å‡ºã—ã‚¨ãƒ©ãƒ¼ã€æ—¢çŸ¥IDã¸ãƒ•ã‚©ãƒ¼ãƒ«ãƒãƒƒã‚¯ã—ã¾ã™")
        return None
    try:
        data = json.loads(r.stdout)
        mv = data.get('modelVersions') or []
        if mv:
            vid = str(mv[0]['id'])
            name = mv[0].get('name','Unknown')
            print(f"âœ… æœ€æ–°ãƒãƒ¼ã‚¸ãƒ§ãƒ³: {name} (ID: {vid})")
            return vid
    except Exception:
        pass
    print("âš ï¸ ãƒ¢ãƒ‡ãƒ«ãƒãƒ¼ã‚¸ãƒ§ãƒ³æƒ…å ±ãŒå–å¾—ã§ããšãƒ•ã‚©ãƒ¼ãƒ«ãƒãƒƒã‚¯ã—ã¾ã™")
    return None

def download_target_model(token: str, version_id: str):
    print(f"ğŸ”„ ãƒ¢ãƒ‡ãƒ«ã‚’ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰/é…ç½®: version_id={version_id}")
    model_dir_local = '/content/stable-diffusion-webui/models/Stable-diffusion'
    os.makedirs(model_dir_local, exist_ok=True)
    output_filename   = "waiNSFWIllustrious_latest.safetensors"
    output_path_drive = f"{drive_models_path}/{output_filename}"
    output_path_local = f"{model_dir_local}/{output_filename}"

    if os.path.exists(output_path_drive) and os.path.getsize(output_path_drive) > 1_000_000:
        print(f"âœ… Driveã«æ—¢å­˜ â†’ ãƒ­ãƒ¼ã‚«ãƒ«ã¸ã‚³ãƒ”ãƒ¼")
        shutil.copy(output_path_drive, output_path_local)
        return

    print("â¬‡ï¸ Civitai ã‹ã‚‰å–å¾—ä¸­...")
    download_url = f"https://civitai.com/api/download/models/{version_id}?token={token}"
    r = subprocess.run([
        'aria2c','-c','-x16','-s16','--check-certificate=false',
        '--summary-interval=10','--content-disposition-default-utf8=true',
        download_url, '-d', drive_models_path, '-o', output_filename
    ], capture_output=True, text=True)
    if r.returncode != 0:
        print("âŒ ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰å¤±æ•—:\n", r.stderr or r.stdout)
        raise SystemExit(1)

    if os.path.exists(output_path_drive) and os.path.getsize(output_path_drive) > 1_000_000:
        shutil.copy(output_path_drive, output_path_local)
        print("âœ… ãƒ¢ãƒ‡ãƒ«é…ç½®å®Œäº†:", output_path_local)
    else:
        print("âŒ Drive ä¸Šã®ãƒ•ã‚¡ã‚¤ãƒ«ãŒè¦‹å½“ãŸã‚‰ãªã„/ä¸å®Œå…¨")
        raise SystemExit(1)

latest_vid = get_latest_model_version_id(os.environ['CIVITAI_API_TOKEN'])
if latest_vid is None:
    print("ğŸ’¡ æ—¢çŸ¥ã®ãƒãƒ¼ã‚¸ãƒ§ãƒ³IDã§ãƒ•ã‚©ãƒ¼ãƒ«ãƒãƒƒã‚¯è©¦è¡Œ")
    for vid in ["1410435","1396035","1378467","1360950"]:
        try:
            download_target_model(os.environ['CIVITAI_API_TOKEN'], vid)
            break
        except SystemExit:
            continue
else:
    download_target_model(os.environ['CIVITAI_API_TOKEN'], latest_vid)

# ----------------------------
# 6) WebUI è¨­å®šãƒ•ã‚¡ã‚¤ãƒ«
# ----------------------------
print("\nâš™ï¸ WebUI è¨­å®šãƒ•ã‚¡ã‚¤ãƒ«ã‚’ä½œæˆ...")
config_content = """{\n  \"samples_save\": true,
  \"samples_format\": \"png\",
  \"grid_save\": true,
  \"grid_format\": \"png\",
  \"enable_pnginfo\": true,
  \"save_selected_only\": true,
  \"jpeg_quality\": 80,
  \"export_for_4chan\": true,
  \"img_downscale_threshold\": 4.0,
  \"target_side_length\": 4000,
  \"img_max_size_mp\": 200,
  \"clean_temp_dir_at_start\": false
}"""
with open('/content/stable-diffusion-webui/config.json','w') as f:
    f.write(config_content)
print("âœ… è¨­å®šå®Œäº†")

# ----------------------------
# 7) ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«ã‚µãƒãƒªãƒ¼
# ----------------------------
print("\n" + "="*60)
print("ğŸ¯ ã‚»ãƒƒãƒˆã‚¢ãƒƒãƒ—å®Œäº†ã‚µãƒãƒªãƒ¼:")
print("="*60)
model_dir = "/content/stable-diffusion-webui/models/Stable-diffusion"
if os.path.exists(model_dir):
    model_files = [f for f in os.listdir(model_dir) if f.endswith(('.safetensors','.ckpt'))]
    print("ğŸ“ ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«ã•ã‚ŒãŸãƒ¢ãƒ‡ãƒ«:")
    if model_files:
        for mf in model_files:
            fp = os.path.join(model_dir, mf)
            try:
                sz = os.path.getsize(fp)/(1024*1024*1024)
                print(f"   âœ… {mf} ({sz:.2f} GB)")
            except Exception:
                print(f"   âœ… {mf}")
    else:
        print("   âŒ ãƒ¢ãƒ‡ãƒ«ãŒè¦‹ã¤ã‹ã‚Šã¾ã›ã‚“")
else:
    print("   âŒ ãƒ¢ãƒ‡ãƒ«ãƒ‡ã‚£ãƒ¬ã‚¯ãƒˆãƒªãŒè¦‹ã¤ã‹ã‚Šã¾ã›ã‚“")

ext_dir = "/content/stable-diffusion-webui/extensions"
if os.path.exists(ext_dir):
    extensions = [d for d in os.listdir(ext_dir) if os.path.isdir(os.path.join(ext_dir, d))]
    print("\nğŸ”Œ ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«ã•ã‚ŒãŸæ‹¡å¼µæ©Ÿèƒ½:")
    for e in extensions: print(f"   âœ… {e}")
print("="*60 + "\n")

# ----------------------------
# 8) èµ·å‹•ç›£è¦–ï¼ˆãƒãƒ¼ãƒˆå´ã§å®Ÿè¡Œï¼‰
# ----------------------------
import threading, requests

def _ping():
    print("ğŸ” èµ·å‹•ç›£è¦–ã‚’é–‹å§‹...")
    for i in range(180):
        try:
            r = requests.get("http://127.0.0.1:7860", timeout=2)
            if r.status_code == 200:
                print("âœ… WebUI èµ·å‹•ã‚’æ¤œå‡º")
                return
        except requests.exceptions.RequestException:
            pass
        if i % 10 == 0:
            print(f"â³ å¾…æ©Ÿä¸­... ({i*2}s)")
        time.sleep(2)
    print("âš ï¸ èµ·å‹•ç›£è¦–ã‚¿ã‚¤ãƒ ã‚¢ã‚¦ãƒˆ")

threading.Thread(target=_ping, daemon=True).start()

# ----------------------------
# 9) WebUI èµ·å‹•ï¼ˆPython 3.10 ã§ï¼‰
#     - A1111ã®å†…éƒ¨venvã‚‚ã€Œpy310ã€ãƒ™ãƒ¼ã‚¹ã§ä½œæˆã•ã‚Œã¾ã™
#     - Torchã¯äº‹å‰ã«å…¥ã‚Œã¦ã„ã‚‹ã®ã§ãã®ã¾ã¾ä½¿ã‚ã‚Œã¾ã™
# ----------------------------
print("\nğŸš€ Stable Diffusion WebUI ã‚’ Python 3.10 ã§èµ·å‹•ã—ã¾ã™")
%cd /content/stable-diffusion-webui

# Torch ã‚³ãƒãƒ³ãƒ‰ã‚’æ˜ç¤ºã—ã¦ãŠãï¼ˆA1111ãŒå¿…è¦æ™‚ã«å‚ç…§ï¼‰
os.environ['TORCH_COMMAND'] = "pip install torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cu121"

# èµ·å‹•ï¼ˆmicromamba çµŒç”±ã§ py310 ã® python ã‚’ä½¿ç”¨ï¼‰
launch_cmd = [
    "micromamba","run","-n","py310","python","launch.py",
    "--share",
    "--listen",
    "--disable-console-progressbars",
    "--opt-sdp-attention",
    "--enable-insecure-extension-access",
    "--lora-dir","/content/stable-diffusion-webui/models/Lora",
    "--medvram-sdxl",
    "--no-half-vae",
    "--api",
    '--cors-allow-origins=*'
]
print("â–¶", " ".join(launch_cmd))
subprocess.Popen(launch_cmd)

print("\nğŸ’¡ ãƒ¡ãƒ¢")
print("- ä»¥é™ã®ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«/å®Ÿè¡Œã¯å¿…ãš `micromamba run -n py310 ...` ã‚’ä½¿ã†ã¨ã€Python 3.10 å´ã§å‹•ãã¾ã™ã€‚")
print("- ãƒãƒ¼ãƒˆãƒ–ãƒƒã‚¯ã® `!python -V` ã¯ 3.11/3.12 ã®ã¾ã¾ãªã®ã§æ··åŒã«æ³¨æ„ã€‚")
```
